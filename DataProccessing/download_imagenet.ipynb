```python
import json
import requests
import time
import shutil
import os.path

base_url = "https://image-net.org/data/winter21_whole/"

with open("./stim_percNatImage_metadata.json") as jf:
    metadata_dict = json.load(jf)

for w, wnid in enumerate(metadata_dict):
    filename = f'{wnid}.tar'
    url = base_url + filename
    local_filename = f'./data/{filename}'

    if os.path.isfile(local_filename):
        print(f'{wnid} already downloaded')
        continue

    with requests.get(url, stream=True) as r:
        if r.status_code != 200:
            print(f'Error: bad status code for {wnid}')
            continue

        filesize_bytes = int(r.headers['content-length'])
        filesize_mb = filesize_bytes / 1024 / 1024

        print(f'[{w}]\t{wnid}\t{filesize_mb:.02f}MB\t', end="")
        # break # debug
        with open(local_filename, 'wb') as f:
            shutil.copyfileobj(r.raw, f)
        
    print(f'✅')

```
